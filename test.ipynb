```python
from langchain_community.tools.tavily_search import TavilySearchResults
search = TavilySearchResults()
```

```python
search.invoke("what is the weather in new delhi?")
```

# Create Retriever (for sim search)

```python
from langchain_community.vectorstores import FAISS
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from PyPDF2 import PdfReader
from langchain_community.vectorstores import FAISS
```

```python
text = ""
reader = PdfReader("Dog-Wikipedia.pdf")
for page in reader.pages:
    text += page.extract_text()
text = text.replace("\n", " ")
print(text)
```

```python
chunks = RecursiveCharacterTextSplitter(chunk_size = 1000, chunk_overlap = 500).split_text(text)
db = FAISS.from_texts(chunks, HuggingFaceEmbeddings())
db.save_local("dog_wikipedia")
```

```python
db = FAISS.load_local("dog_wikipedia", HuggingFaceEmbeddings(), allow_dangerous_deserialization=True)
retriever = db.as_retriever()
```

```python
retriever.invoke("eating dogs")
```

# Creating tools

```python
from langchain.tools.retriever import create_retriever_tool

retriever_tool = create_retriever_tool(
    retriever,
    "dog_search",
    "Search for information about Dogs. For any questions about dogs, you must use this tool!",
)
```

```python
tools = [search, retriever_tool]
```

# Creating the agent

```python
!pip install groq
!pip install python-dotenv
!pip install langchain-groq
```

```python
from langchain_groq import ChatGroq
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
llm = ChatGroq(
    api_key = os.getenv("GROQ_API_KEY"),
    model = "llama3-70b-8192"
)
```

```python
!pip install langchainhub
from langchain import hub

# Get the prompt to use - you can modify this!
prompt = hub.pull("hwchase17/openai-functions-agent")
prompt.messages
```

```python
from langchain.agents import create_tool_calling_agent

agent = create_tool_calling_agent(llm, tools, prompt)
```

```python
from langchain.agents import AgentExecutor

agent_executor = AgentExecutor(agent=agent, tools=tools, verbose=True)
```

# Run the agent

```python
agent_executor.invoke({"input": "hi!"})
```

```python
agent_executor.invoke({"input": "what is the weather in new delhi?"})
```

```python
agent_executor.invoke({"input": "what do the dogs eat?"})
```
